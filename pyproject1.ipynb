# Project-1

**Step-1:Loading Dataset**

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Loading the dataset

file_path = 'schooldata.csv'
data = pd.read_csv(file_path)

#displaying first few rows

data.head()
```

**Step-2:Data Cleaning**

```python
# Replace 's' with NaN and convert relevant columns to numeric
data.replace('s', pd.NA, inplace=True)

# Convert numeric columns to  data types(Defining the columns that should be numeric)
numeric_columns = ['Mean Score', 'Total Tested', 'Number Scoring Below 65', 'Percent Scoring Below 65',
                   'Number Scoring 65 or Above', 'Percent Scoring 65 or Above',
                   'Number Scoring 80 or Above', 'Percent Scoring 80 or Above',
                   'Number Scoring CR', 'Percent Scoring CR']
for column in numeric_columns:
    data[column] = pd.to_numeric(data[column], errors='coerce')


#DATA CLEANNG-TYPE-2
#Handle missing values in numeric columns by filling with the mean or median
#This helps in maintaining the data integrity and allows for better analysis
for column in numeric_columns:
    if data[column].isnull().sum() > 0:
        if data[column].dtype == 'float64':
            data[column].fillna(data[column].mean(), inplace=True)
        else:
            data[column].fillna(data[column].median(), inplace=True)


#DATA CLEANNG-TYPE-3
# Drop rows with any remaining missing values
data.dropna(inplace=True)

# for Verification
data.info()
data.describe()
```

**Step-3:Comparison-1**

```python
# Filter data for the "P.S. 034 Franklin D. Roosevel"
chosen_school = data[data['School Name'] == 'P.S. 034 Franklin D. Roosevelt']

# Descriptive statistics for the above school
mean_score_stats_chosen = chosen_school['Mean Score'].describe()
print("Descriptive Statistics for P.S. 034 Franklin D. Roosevelt:")
print(mean_score_stats_chosen)

# Descriptive statistics for all schools
mean_score_stats_overall = data['Mean Score'].describe()
print("\nDescriptive Statistics for All Schools:")
print(mean_score_stats_overall)

# Descriptive statistics for schools in a  'Manhattan' borough
borough_schools = data[data['School DBN'].str.startswith('01')]
mean_score_stats_borough = borough_schools['Mean Score'].describe()
print("\nDescriptive Statistics for Manhattan Schools:")
print(mean_score_stats_borough)
```

**`Another Comparison`**

```python
# Comparison of another chosen school'P.S. 140 Nathan Straus' with other school
other_school = data[data['School Name'] == 'P.S. 140 Nathan Straus']
mean_score_stats_other = other_school['Mean Score'].describe()
print("\nDescriptive Statistics for P.S. 140 Nathan Straus:")
print(mean_score_stats_other)
```

**Step-4:Plotting Results**

```python
# Box plot comparison
plt.figure(figsize=(12, 8))
plt.boxplot([chosen_school['Mean Score'].dropna(), borough_schools['Mean Score'].dropna(), data['Mean Score'].dropna()],
            labels=['P.S. 034 Franklin D. Roosevelt', 'Manhattan Schools', 'All Schools'])
plt.title('Comparison of Mean Scores')
plt.ylabel('Mean Score')
plt.show()

# Histogram of Mean Scores for Chosen School
plt.figure(figsize=(12, 8))
plt.hist(chosen_school['Mean Score'].dropna(), bins=10, alpha=0.7, label='P.S. 034 Franklin D. Roosevelt')
plt.hist(borough_schools['Mean Score'].dropna(), bins=10, alpha=0.7, label='Manhattan Schools')
plt.hist(data['Mean Score'].dropna(), bins=10, alpha=0.7, label='All Schools')
plt.title('Histogram of Mean Scores')
plt.xlabel('Mean Score')
plt.ylabel('Frequency')
plt.legend()
plt.show()

# Scatter plot of Total Tested vs. Mean Score
plt.figure(figsize=(12, 8))
plt.scatter(chosen_school['Total Tested'], chosen_school['Mean Score'], label='P.S. 034 Franklin D. Roosevelt')
plt.scatter(borough_schools['Total Tested'], borough_schools['Mean Score'], label='Manhattan Schools')
plt.scatter(data['Total Tested'], data['Mean Score'], label='All Schools')
plt.title('Total Tested vs. Mean Score')
plt.xlabel('Total Tested')
plt.ylabel('Mean Score')
plt.legend()
plt.show()
```

# Conclusion

> Based on the analysis, P.S. 034 Franklin D. Roosevelt shows a higher mean score compared to the average mean score of schools in Manhattan and all schools in the dataset.

This indicates better student performance in the chosen school. The visualizations further support these findings, showing clear distinctions in the distributions of mean scores.
